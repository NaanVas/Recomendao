# **NMF**

```python
def nmf(k, R_train, R_val, num_epochs, learning_rate):
    num_users = R_train.shape[0]
    num_items = R_train.shape[1]

    W = torch.rand(num_users, k, requires_grad=True)
    H = torch.rand(k, num_items, requires_grad=True)

    loss_func = torch.nn.MSELoss()
    optimizer = torch.optim.SGD([W, H], lr=learning_rate, momentum=0.9)

    best_val_loss = float('inf')
    best_W = None
    best_H = None

    for epoch in tqdm(range(num_epochs), "Epoca: "):
        # Forward pass
        R_pred = torch.mm(W, H)

        R_pred_train = R_pred[:R_train.shape[0], :R_train.shape[1]]
        R_pred_val = R_pred[:R_val.shape[0], :R_val.shape[1]]

        train_loss = loss_func(R_pred_train, R_train)
        #train_mse = train_loss.item()
        #print("MSE Train:", train_mse)
        val_loss = loss_func(R_pred_val, R_val)
        #val_mse = val_loss.item()
        #print("MSE Validação: ", val_mse)

        # Backward pass and optimization
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_W = W.clone().detach()
            best_H = H.clone().detach()

    return best_W, best_H
```

```python
# Dividindo o conjunto de dados em treino, validação e teste (70% treino, 15% validação, 15% teste)
train_data, test_data = train_test_split(df_np, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(test_data, test_size=0.5, random_state=42)

# Convertendo os dados em tensores do PyTorch
train_tensor = torch.tensor(train_data, dtype=torch.float32)
val_tensor = torch.tensor(val_data, dtype=torch.float32)
test_tensor = torch.tensor(test_data, dtype=torch.float32)
```

```python
k = 20
best_mse = float('inf')
num_epochs = 10000
learning_rate = 0.01
```

```python
W, H = nmf(k, train_tensor, val_tensor, num_epochs=num_epochs, learning_rate=learning_rate)
    
# Testando o melhor modelo no conjunto de teste
R_pred_test = torch.mm(W, H)
R_pred_test_selected = R_pred_test[:test_tensor.shape[0], :test_tensor.shape[1]]

mse_test = torch.nn.functional.mse_loss(R_pred_test_selected, test_tensor).item()

print(f"MSE no conjunto de teste: {mse_test}")
```

```python
threshold_value = 2

# Converta os tensores para arrays NumPy
R_pred_test_selected_np = R_pred_test_selected.cpu().detach().numpy()

# Converta as previsões para classes binárias
predicted_classes = (R_pred_test_selected_np > threshold_value).astype(int)
true_classes = (test_data > threshold_value).astype(int)

# Calcule as métricas
accuracy = accuracy_score(true_classes.ravel(), predicted_classes.ravel())
precision = precision_score(true_classes.ravel(), predicted_classes.ravel())
recall = recall_score(true_classes.ravel(), predicted_classes.ravel())
f1 = f1_score(true_classes.ravel(), predicted_classes.ravel())

# Exiba as métricas calculadas
print(f"Acurácia: {accuracy}")
print(f"Precisão: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

# **SVD**

```python
def svd(k, R_train, R_val, num_epochs, learning_rate):
    num_users = R_train.shape[0]
    num_items = R_train.shape[1]

    # Inicializar matrizes de fatorização aleatórias
    U = torch.rand(num_users, k, requires_grad=True)
    S = torch.rand(k, requires_grad=True)
    V = torch.rand(k, num_items, requires_grad=True)

    loss_func = torch.nn.MSELoss()
    optimizer = torch.optim.SGD([U, S, V], lr=learning_rate, momentum=0.9)

    best_val_loss = float('inf')
    best_U = None
    best_S = None
    best_V = None

    for epoch in tqdm(range(num_epochs), "Epoch: "):
        # Forward pass
        R_pred = torch.matmul(torch.matmul(U, torch.diag(S)), V)

        R_pred_train = R_pred[:R_train.shape[0], :R_train.shape[1]]
        R_pred_val = R_pred[:R_val.shape[0], :R_val.shape[1]]

        train_loss = loss_func(R_pred_train, R_train)
        #train_mse = train_loss.item()
        #print("MSE Train:", train_mse)
        val_loss = loss_func(R_pred_val, R_val)
        #val_mse = val_loss.item()
        #print("MSE Validação: ", val_mse)

        # Backward pass and optimization
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_U = U.clone().detach()
            best_S = S.clone().detach()
            best_V = V.clone().detach()

    return best_U, best_S, best_V
```

```python
U, S, V = svd(k, train_tensor, val_tensor, num_epochs=num_epochs, learning_rate=learning_rate)
    
# Testando o melhor modelo no conjunto de teste
R_pred_test = torch.mm(torch.mm(U, torch.diag(S)), V)
R_pred_test_selected = R_pred_test[:test_tensor.shape[0], :test_tensor.shape[1]]

mse_test = torch.nn.functional.mse_loss(R_pred_test_selected, test_tensor).item()

print(f"MSE no conjunto de teste: {mse_test}")
```
